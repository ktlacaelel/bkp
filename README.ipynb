# What is it?

A shell tool to manage backups that uses aws s3 buckets as storage.

# What does it do?
It ensures all backups are stored in the same fasion.
1. The backup is compressed.
1. The backup is compressed into 1 file.
1. The backup is encapsulated in 1 dir, so files won't spread all over the place during extraction.

### Writes a manifest in json format
The manifest contains the following fields.
1. Summary - Brief oneliner that tells what the backup is about.
1. Background - A place for context about the backup.
1. Ticket - The ticket in relation to the backup.
1. Owner - Who performed the backup.
1. Date - Automatically Assigns a date but flexible for manual modification.

### Shell Tool
The shell tool comes with some features.
#### Need a backup?

1. Install the gem.
1. List backups (organized by date)
1. Review details of the backup in a human friendly format.
1. Download the backup with very few keystrokes!

#### Need to create one?

1. Install the gem.
1. Generate a yaml config that has all the fields for the manifest (and directory to backup).
1. Validate the yaml with the *check* command.
1. Upload your backup.
1. Review your upload.
1. Download and review the data is there (optional)

#### Got a question about a backup?

1. Install the gem.
1. Search the backups.
1. Review details of the backup in a human friendly format.
1. Download the backup and review the data.

# Configuration

Run the `bkp` command on your terminal it will ask you two things:

1. What is the AWS S3 `bucket` where you want to store your backups.
1. What is the `path` inside the bucket where the backups will live.

Example:

s3://`[BUCKET]`/`[PATH]`/`[YOUR BACKUPS WILL LIVE HERE]`
